```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
```

```python
temp = []
with open('hist-lena-2.txt') as f:
    cur = f.readlines()
    for num in cur:
        temp.append(int(num[:-1]))
```

```python
print(temp)
```

```python
x = []
x = [0,127]
plt.figure()
#plt.bar(x,temp, width=2)
plt.plot(x, temp)
plt.xlim(0, 255)
plt.title('Histogram of Lena with Bin=2')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()
```

```python
temp = []
with open('hist-lena-128.txt') as f:
    cur = f.readlines()
    for num in cur:
        temp.append(int(num[:-1]))
pix = []
for i in range(256):
    if i%2 != 0:
        pix.append(0)
    else:
        pix.append(temp[int(i/2)])
x = [i for i in range(256)]
plt.figure()
plt.bar(x,pix, width=2)
plt.xlim(0, 255)
plt.title('Histogram of Lena with Bin=128')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
```

```python
with open('hist-part3-D.txt') as f:
    lines = f.read().splitlines()
    print(lines)
```

```python
orig = []
orig = lines[1:257]
new = []
new = lines[258:514]
x = [i for i in range(256)]

plt.figure()
plt.bar(x,orig, width=10)
plt.xlim(0, 255)
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.bar(x,new, width=5)
plt.xlim(0, 255)
plt.title('Histogram of Normalized D(orange) and Original D(blue) with Bin=256')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.plot()
```

```python
with open('hist-part3-Lena.txt') as f:
    lines = f.read().splitlines()
    print(lines)
```

```python
orig_l = []
for i in range(256):
    orig_l.append(int(lines[i+1]))
print(orig_l)
new_l = []
for i in range(256):
    new_l.append(int(lines[258+i]))
x = [i for i in range(256)]

plt.figure()
plt.bar(x,orig_l, width=1)
plt.bar(x,new_l, width=1)
plt.xlim(0, 255)
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.title('Histogram of Original Lena(blue) and Normalized Lena(orange) with Bin=256')
```

```python
plt.figure()
plt.bar(x,new_l, width=1)
plt.xlim(0, 255)
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.title('Histogram of Normalized Lena with Bin=256')
```
